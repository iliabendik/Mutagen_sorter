```python
from audiofile import bit_depth
import mutagen
import numpy as np
import pandas as pd
import os
import PySimpleGUI as sg
import re
from mutagen.mp3 import MP3
import audiofile as af
from mutagen.wave import WAVE
from mutagen.flac import FLAC
from mutagen.dsf import DSF
from shutil import move
import pandas as pd
```

```python
def checking_dir(path, files_list, path_to_move, a1, a2, a3, a4):
    name, length, size, channels_quantity, bitrate, samplerate, bitdepth, compression, quality = [], [], [], [], [], [], [], [], []
    count_mp3, count_wav, count_flac, count_dsf = 0, 0, 0, 0
    os.chdir(path)
    for g in files_list:
        if g.endswith('.mp3') and a1:
            count_mp3+=1
            file = MP3(g)
            print('Название файла:', g)
            print('Длина файла:', file.info.length, 'с')
            print('Размер файла:', os.stat(g).st_size, 'байт')
            print('Количество каналов:', file.info.channels)
            print('Битрейт:', file.info.bitrate)
            print('Частота дискретизации:', file.info.sample_rate, 'Гц')
            name.append(g)
            length.append(file.info.length)
            size.append(os.stat(g).st_size)
            channels_quantity.append(file.info.channels)
            bitrate.append(file.info.bitrate)
            samplerate.append(file.info.sample_rate)
            bitdepth.append(None)
            compression.append(None)
            
            if file.info.bitrate < 90000:
                print('Низкое качество\n')
                quality.append('Низкое качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\Low_Quality\\' + g)
            elif file.info.bitrate >= 90000 and file.info.bitrate < 128000:
                print('Среднее качество\n')
                quality.append('Среднее качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\Medium_Quality\\' + g)
            elif file.info.bitrate >= 128000:
                print('Высокое качество\n')
                quality.append('Высокое качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\High_Quality\\' + g)
                
        elif g.endswith('.wav') and a2:
            count_wav+=1
            file = WAVE(g)
            print('Название файла:', g)
            print('Длина файла:', file.info.length, 'с')
            print('Размер файла:', os.stat(g).st_size, 'байт')
            print('Количество каналов:', file.info.channels)
            print('Битрейт:', file.info.bitrate)
            print('Частота дискретизации:', file.info.sample_rate, 'Гц')
            print('Разрядность:', bit_depth(g), 'бит')
            name.append(g)
            length.append(file.info.length)
            size.append(os.stat(g).st_size)
            channels_quantity.append(file.info.channels)
            bitrate.append(file.info.bitrate)
            samplerate.append(file.info.sample_rate)
            bitdepth.append(bit_depth(g))
            compression.append(None)
            
            if file.info.sample_rate * bit_depth(g) * file.info.channels < 90000:
                print('Низкое качество\n')
                quality.append('Низкое качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\Low_Quality\\' + g)
            elif file.info.sample_rate * bit_depth(g) * file.info.channels >= 90000 and file.info.sample_rate * bit_depth(g) * file.info.channels < 128000:
                print('Среднее качество\n')
                quality.append('Среднее качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\Medium_Quality\\' + g)
            elif file.info.sample_rate * bit_depth(g) * file.info.channels >= 128000:
                print('Высокое качество\n')
                quality.append('Высокое качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\High_Quality\\' + g)
            
        elif g.endswith('.flac') and a3:
            count_flac+=1
            file = FLAC(g)
            print('Название файла:', g)
            print('Длина файла:', file.info.length, 'с')
            print('Размер файла:', os.stat(g).st_size, 'байт')
            print('Количество каналов:', file.info.channels)
            print('Битрейт:', file.info.bitrate)
            print('Частота дискретизации:', file.info.sample_rate, 'Гц')
            print('Разрядность:', bit_depth(g), 'бит')
            print('Уровень сжатия:', file.info.sample_rate*bit_depth(g)*file.info.channels/file.info.bitrate)
            name.append(g)
            length.append(file.info.length)
            size.append(os.stat(g).st_size)
            channels_quantity.append(file.info.channels)
            bitrate.append(file.info.bitrate)
            samplerate.append(file.info.sample_rate)
            bitdepth.append(bit_depth(g))
            compression.append(file.info.sample_rate*bit_depth(g)*file.info.channels/file.info.bitrate)
            
            if file.info.sample_rate * bit_depth(g) * file.info.channels / file.info.bitrate < 90000:
                print('Низкое качество\n')
                quality.append('Низкое качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\Low_Quality\\' + g)
            elif file.info.sample_rate * bit_depth(g) * file.info.channels / file.info.bitrate >= 90000 and file.info.sample_rate*bit_depth(g)*file.info.channels/file.info.bitrate < 128000:
                print('Среднее качество\n')
                quality.append('Среднее качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\Medium_Quality\\' + g)
            elif file.info.sample_rate * bit_depth(g) * file.info.channels / file.info.bitrate >= 128000:
                print('Высокое качество\n')
                quality.append('Высокое качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\High_Quality\\' + g)
            
        elif g.endswith('.dsf') and a4:
            count_dsf+=1
            file = DSF(g)
            print('Название файла:', g)
            print('Длина файла:', file.info.length, 'с')
            print('Размер файла:', os.stat(g).st_size, 'байт')
            print('Количество каналов:', file.info.channels)
            print('Битрейт:', file.info.bitrate)
            print('Частота дискретизации:', file.info.sample_rate, 'Гц')
            print('Разрядность:', bit_depth(g), 'бит')
            name.append(g)
            length.append(file.info.length)
            size.append(os.stat(g).st_size)
            channels_quantity.append(file.info.channels)
            bitrate.append(file.info.bitrate)
            samplerate.append(file.info.sample_rate)
            bitdepth.append(bit_depth(g))
            compression.append(None)
            
            if file.info.bitrate < 90000:
                print('Низкое качество\n')
                quality.append('Низкое качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\Low_Quality\\' + g)
            elif file.info.bitrate >= 90000 and file.info.bitrate < 128000:
                print('Среднее качество\n')
                quality.append('Среднее качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\Medium_Quality\\' + g)
            elif file.info.bitrate >= 128000:
                print('Высокое качество\n')
                quality.append('Высокое качество')
                move(os.getcwd() + '\\' + g, path_to_move + '\\High_Quality\\' + g)
            
    return [count_mp3, count_wav, count_flac, count_dsf], [name, length, size, channels_quantity, bitrate, samplerate, bitdepth, compression, quality]
```

```python
layout = [
    [sg.Text('Search Pathway'), sg.InputText(), sg.FolderBrowse()],
    [sg.Text('Destination Pathway'), sg.InputText(), sg.FolderBrowse()],
    [sg.Checkbox('MP3'), sg.Checkbox('WAV'), sg.Checkbox('FLAC'), sg.Checkbox('DSF')],
    [sg.Checkbox('Вложенные папки')],
    [sg.Output(size=(88, 20))],
    [sg.Submit(), sg.Cancel()]]
window = sg.Window('Getting info', layout)
while True:
    
    event, values = window.read()
    if event in (None, 'Exit', 'Cancel'):
        break
    if event == 'Submit':
        os.chdir(values[1])
        os.mkdir('Low_Quality')
        os.mkdir('Medium_Quality')
        os.mkdir('High_Quality')
        
        name, length, size, channels_quantity, bitrate, samplerate, bitdepth, compression, quality = [], [], [], [], [], [], [], [], []

        count_mp3, count_wav, count_flac, count_dsf = 0, 0, 0, 0
        
        if values[6]:
            everything = list(os.walk(values[0], topdown=True))
            for i in range(len(everything)):
                everything[i] = list(everything[i])
                everything[i].pop(1)
            for item in everything:
                              
                addition, params = checking_dir(item[0], item[1], values[1], values[2], values[3], values[4], values[5])
                count_mp3+=addition[0]
                count_wav+=addition[1]
                count_flac+=addition[2]
                count_dsf+=addition[3]
                
                name += params[0]
                length += params[1]
                size += params[2]
                channels_quantity += params[3]
                bitrate += params[4]
                samplerate += params[5]
                bitdepth += params[6]
                compression += params[7]
                quality += params[8]
                
        else:
            os.chdir(values[0])
            addition, params = checking_dir(os.getcwd(), os.listdir(), values[1], values[2], values[3], values[4], values[5])
            count_mp3+=addition[0]
            count_wav+=addition[1]
            count_flac+=addition[2]
            count_dsf+=addition[3]
            
            name += params[0]
            length += params[1]
            size += params[2]
            channels_quantity += params[3]
            bitrate += params[4]
            samplerate += params[5]
            bitdepth += params[6]
            compression += params[7]
            quality += params[8]
            
        if values[2] or values[3] or values[4] or values[5]:
            print('Количество файлов:')
        if values[2]:
            print('MP3 -', count_mp3)
        if values[3]:
            print('WAV -', count_wav)
        if values[4]:
            print('FLAC -', count_flac)
        if values[5]:
            print('DSF -', count_dsf)
    print('\n')
window.close()
pd.DataFrame({'Название': name, 'Длительность, с': length, 'Размер, байт': size, 'Количество каналов': channels_quantity, 'Битрейт': bitrate, 'Частота дискретизации, Гц': samplerate, 'Разрядность, бит': bitdepth, 'Уровень сжатия': compression, 'Качество': quality})
```
